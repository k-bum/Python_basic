```python
import numpy as np

x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

x_2 = range(100)
y_2 = [np.sin(value) for value in x_2]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)
ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(x_1, y_1, c = 'b')
ax_2.plot(x_2, y_2, c = 'g')
plt.show()
```

```python
x_1 = range(100)
y_1 = [value for value in x_1]

x_2 = range(100)
y_2 = [value + 100 for value in x_2]

plt.plot(x_1, y_1, color = '#000000')
plt.plot(x_2, y_2, c = 'c')

plt.show()
```

```python
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed')
plt.plot(x_2, y_2, c = 'r', ls = 'dotted')

plt.title("Two lines")
plt.show()
```

```python
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed')
plt.plot(x_2, y_2, c = 'g', linestyle = 'dotted')

plt.title('$y = ax + b$')
plt.xlabel('$x_line$')
plt.ylabel('$y_line$')
```

```python
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed')
plt.plot(x_2, y_2, c = 'g', linestyle = 'dotted')

plt.text(50, 70, 'Line_1')
plt.annotate('Line_2', xy = (50, 150), xytext = (20, 175), 
            arrowprops = dict(facecolor = 'black', shrink = 0.05))

plt.title('$y = ax + b$')
plt.xlabel('$x_line$')
plt.ylabel('$y_line$')

plt.show()
```

```python
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed', label = 'Line_1')
plt.plot(x_2, y_2, c = 'g', linestyle = 'dotted', label = 'Line_2')
plt.legend(shadow = True, fancybox = True, loc = 'upper right')

plt.text(50, 70, 'Line_1')
plt.annotate('Line_2', xy = (50, 150), xytext = (20, 175), 
            arrowprops = dict(facecolor = 'black', shrink = 0.05))

plt.title('$y = ax + b$')
plt.xlabel('$x_line$')
plt.ylabel('$y_line$')

plt.show()
```

```python
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed', label = 'Line_1')
plt.plot(x_2, y_2, c = 'g', linestyle = 'dotted', label = 'Line_2')
plt.legend(shadow = True, fancybox = True, loc = 'upper right')

plt.grid(True, lw = 0.4, ls = '--', c = '.90')
plt.xlim(-100, 200)
plt.ylim(-100, 200)

plt.title('$y = ax + b$')
plt.xlabel('$x_line$')
plt.ylabel('$y_line$')

plt.show()
```

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

x = np.arange(0,8,2)
```

```python
data = np.array([[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
X
```

```python
data = np.array([[5., 5., 5., 5.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), 
            color = color_list[i], label=data_label[i])
plt.legend()
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```
